# Multi-modal Models

## What is Multimodality? 📸📝🎵

A *modality* means a medium or a way in which something exists or is done.

We use our 5 sense organs to recieve sensory inputs in multiple ways (modalities):

1. 👀 eyes to see
2. 👂️ ears to hear
3. 🤝 skin to touch
4. 👃 nose to smell
5. 👅 tongue to taste

### Real life examples

- 🖼️ **Vision + Text**: Infographics, Memes, Articles, Blogs.
- 🗣️ **Vision + Audio**: In-person or online conversations.
- 🎬 **Vision + Audio + Text**: Watching YouTube videos or movies with captions, social media content in general is multimodal.
- 🎵 **Audio + Text**: Voice notes, songs with lyrics.

### Multimodal Datasets

A dataset consisting of multiple modalities is a multimodal dataset. Out of the common modality combinations let us see some examples:

- Vision + Text: [Visual Storytelling Dataset](https://visionandlanguage.net/VIST/), [Visual Question Answering Dataset](https://visualqa.org/download.html), [LAION-5B Dataset](https://laion.ai/blog/laion-5b/).
- Vision + Audio: [VGG-Sound Dataset](https://www.robots.ox.ac.uk/~vgg/data/vggsound/), [RAVDESS Dataset](https://zenodo.org/records/1188976), [Audio-Visual Identity Database (AVID)](https://www.avid.wiki/Main_Page).
- Vision + Audio + Text: [RECOLA Database](https://diuf.unifr.ch/main/diva/recola/), [IEMOCAP Dataset](https://sail.usc.edu/iemocap/).

## Visual Question Answering Models

VQA models have two parts:

1. vision encoder to "see"
2. text encoder to "read"

![](../assets/vlm_pre-trained_models.png)

### Training Objective

The training objective: learn the vision-language correlation of the embeddings of the two, such that each one is representative of the other. In other words, if you mask the word "dog", the picture region should help the model predict it. Likewise, if you mask the region in the image of the face of the dog, the word "dog" should help the model predict it. This is done by correlating the embedding vectors of the picture-description pair.

Given enough data, models learn what's a "dog", and if replaced by a "cat", what should the image look like instead, and vice-versa.

## Document Visual Question Answering (DocVQA)


* **Task:** [Document Question Answering](https://huggingface.co/models?pipeline_tag=document-question-answering)- Analyze visual and textual data, infer relationships, and generate a natural language response.
* **Input:** Document image and text question.
* **Output:** Accurate text answer.

![](../assets/vlm_input_form.png)

Representative VQA Models:

* [**impira/layoutlm-document-qa**](https://huggingface.co/impira/layoutlm-document-qa): Approximately **113 million** parameters. It is based on the LayoutLM (base) architecture, which incorporates text, layout, and image information.
* [**naver-clova-ix/donut-base-finetuned-docvqa**](https://huggingface.co/naver-clova-ix/donut-base-finetuned-docvqa): Approximately **201 million** parameters. This model uses a Swin Transformer encoder and a mBART decoder for OCR-free document processing.
* [**facebook/nougat-base**](https://huggingface.co/facebook/nougat-base): Approximately **250 million** parameters. It is a transformer-based encoder-decoder model specifically optimized for transcribing scientific documents into markup.

Learn more about how to train and use DocVQA models in HuggingFace `transformers` library [here](https://huggingface.co/docs/transformers/tasks/document_question_answering).

### GLM-OCR

> [GLM-OCR](https://huggingface.co/zai-org/GLM-OCR) is a multimodal OCR model for complex document understanding, built on the GLM-V encoder–decoder architecture. It introduces Multi-Token Prediction (MTP) loss and stable full-task reinforcement learning to improve training efficiency, recognition accuracy, and generalization. The model integrates the CogViT visual encoder pre-trained on large-scale image–text data, a lightweight cross-modal connector with efficient token downsampling, and a GLM-0.5B language decoder. Combined with a two-stage pipeline of layout analysis and parallel recognition based on PP-DocLayout-V3, GLM-OCR delivers robust and high-quality OCR performance across diverse document layouts.

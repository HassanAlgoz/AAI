## Natural Language Processing (NLP)

NLP bridges the gap between computers and human language. It combines elements of:

1. computer science, 
2. linguistics, 
3. artificial intelligence

Recent success in NLP has been driven by **Deep Learning approach** which is based on [massive datasets](https://github.com/niderhoff/nlp-datasets) and models that can learn from data.

### NLU & NLG

**Natural Language Understanding (NLU)**: extract meaning, intention, emotion, importance, and correlation between words/texts/speech.

- Examples: Classification, NER, and ASR

**Natural Language Generation (NLG)**: generate the most likely sequence given previous sequence.

- Examples: Question Answering, Summarization, Translation, and Coding.

## 1. Text Classification

**Text Classification** is the process of assigning a category or label to a piece of text.

### 1.1 Sentiment Analysis

**Sentiment Analysis** identifies the emotional tone of text.

* **Positive:** "This coffee is amazing!"
* **Negative:** "My order arrived cold and late."
* **Neutral:** "The store opens at 8 AM."

If we don't specify a model, a default model will be used:

```python
from transformers import pipeline

classifier = pipeline(task="sentiment-analysis")
```

```python
output = classifier([
    "I've been waiting for a HuggingFace course my whole life.",
    "I hate this so much!"
])

print(output)
```

```python
if output[0]['label'] == 'POSITIVE':
    print("the user likes the product")
```

### 1.3 Zero-shot Classification

Models trained with NLI can do zero-shot classification; i.e., rather than being specifically trained on a fixed set of classes, they can take `candidate_labels` at inference time, and be able to output class probability on them:

```python
from transformers import pipeline

zero_shot_classifier = pipeline("zero-shot-classification")
```

```python
output = zero_shot_classifier(
    "This is a course about the Transformers library",
    candidate_labels=["education", "politics", "business"],
)

print(output)
```

```python
output = zero_shot_classifier(
    "I need my money back",
    candidate_labels=[
        "Refund",
        "Question",
        "Complain"
    ],
)

label = output['labels'][0]

if label == 'Refund':
    print("Trigger workflow for getting a refund")
elif label == 'Complain':
    print("Escalate to customer support manager")
elif label == 'Question':
    print("Direct user to the FAQ documentation")
else:
    print(f"Categorized as: {label}")
```

## 3. Named-entity Recognition (NER)

**Named entity recognition (NER)** aims to extract entities in a piece of text into predefined categories such as: **personal names**, **organizations**, **locations**, and **quantities**.

Task name: `token-classification`.

![](../assets/NER.png){fig-align="center" .r-stretch}

Example: [GLiNER](https://huggingface.co/NAMAA-Space/gliner_arabic-v2.1).

### Applications

- Building knowledge graphs for Arabic content.
- Enhancing search and recommendation systems with entity-aware features.


```python
from transformers import pipeline

ner = pipeline(
    task="token-classification",
    aggregation_strategy="simple"
)
ner("My name is Sylvain and I work at Hugging Face in Brooklyn.")
```

```python
from gliner import GLiNER

# Load the model
model = GLiNER.from_pretrained("NAMAA-Space/gliner_arabic-v2.1")

# Example text (Arabic)
text = "زار أحمد شركة البيانات الوطنية في الرياض يوم الاثنين."
labels = ["شخص", "منظمة", "تاريخ", "موقع"]

# Perform entity prediction
entities = model.predict_entities(text, labels, threshold=0.5)

# Display predicted entities and their labels
for entity in entities:
    print(f"Entity: {entity['text']} | Label: {entity['label']} | Score: {entity['score']:.3f}")
```

## 7. Text Generation (LLM)

LLM [.generate()](https://huggingface.co/docs/transformers/v5.5.0/en/main_classes/text_generation#transformers.GenerationMixin.generate) method auto-regressively generates the next sequence given some initial `prompt`. Generation can be controlled:

```python
from transformers import pipeline, GenerationConfig

generation_config = GenerationConfig(
    max_new_tokens=50,
    temperature=1.5
)

generator = pipeline(
    task="text-generation",
    model="HuggingFaceTB/SmolLM2-360M",
    generation_config=generation_config,
)
```

### Common Generation Options

See: [Common Options](https://huggingface.co/docs/transformers/v5.5.0/en/llm_tutorial#common-options):

- **`max_new_tokens` (`int`)**: Controls the maximum number of tokens generated. Make sure to set it explicitly—defaults are usually small.
- **`num_beams` (`int`)**: Activates beam search when set above 1. Useful for input-grounded tasks. See [this guide](https://huggingface.co/docs/transformers/v5.5.0/en/generation_strategies).
- **`repetition_penalty` (`float`)**: Set to greater than 1.0 to discourage the model from repeating itself.
- **`temperature` (`float`)**: Influences randomness of generation. High values (`>0.8`) make outputs more creative; low values (`<0.4`) make them more focused and predictable.

```python
output = generator(
    "In this course, we will teach you how to",
    max_new_tokens=30,
    num_return_sequences=3,
)
```

## LLM for Specific Tasks

```python
from transformers import pipeline

checkpoint = "HuggingFaceTB/SmolLM2-360M"
generator = pipeline(
    "text-generation",
    model=checkpoint,
    device="auto"
)


def emotion_classifier(batch: list[str], emotions: list[str]) -> list[str]:
    """
    Classifies the emotion of each text.
    Accepts a list of strings (batch).
    """
    emotion_options = ", ".join(emotions)
    prompts = [
        (
            f"Select the primary emotion expressed in the following text from the provided list. Respond with only one emotion.\n\n"
            f"Possible emotions: {emotion_options}\n"
            f"Text:\n{t}\n\nEmotion:"
        )
        for t in batch
    ]
    results = generator(prompts, max_new_tokens=8, do_sample=False)
    outputs = [r["generated_text"] for r in results]
    return outputs


def question_answerer(batch: list[str]) -> list[str]:
    """
    Answers questions about the provided texts using text generation.
    Accepts a list of strings (batch).
    """
    prompts = [
        (
            "Given the following text, provide a concise answer to the main question or information presented.\n\n"
            f"Text:\n{t}\n\nAnswer:"
        )
        for t in batch
    ]
    results = generator(prompts, max_new_tokens=20, do_sample=False)
    outputs = [r["generated_text"] for r in results]
    return outputs


def summarizer(texts: list[str], summary_length: int = 40) -> list[str]:
    """
    Summarizes the input texts.
    Accepts a list of strings (batch).
    """
    prompts = [
        (
            "Read the following text and write a clear, concise summary that captures the main points.\n\n"
            f"Text:\n{t}\n\nSummary:"
        )
        for t in texts
    ]
    results = generator(prompts, max_new_tokens=summary_length, do_sample=False)
    outputs = [r["generated_text"] for r in results]
    return outputs


def translator(texts: list[str], target_language: str = "Arabic") -> list[str]:
    """
    Translates texts to the given target language.
    Accepts a list of strings (batch).
    """
    prompts = [
        (
            f"Translate the following text into {target_language}. Provide only the translated text, and do not include any explanation.\n\n"
            f"Text:\n{t}\n\nTranslation:"
        )
        for t in texts
    ]
    results = generator(prompts, max_new_tokens=40, do_sample=False)
    outputs = [r["generated_text"] for r in results]
    return outputs
```

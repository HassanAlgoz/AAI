```python
import pandas as pd
import sklearn
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error
import numpy as np


```

### Step 1. Load the data

```python
salary_df = pd.read_csv("https://raw.githubusercontent.com/HassanAlgoz/B5/main/content/datasets/Salary%20Data.csv")
salary_df.head()
```

```python
salary_df.columns
```

### Step 2.a Assign variables `X` to the features and `y` to the target

```python
x= salary_df[['Experience Years']]
y= salary_df['Salary']
```

### Step 2.b print the type of each

```python
print(type(x))
print(type(y)) 
```

### Step 2.c identify whether the target is categorical or numerical & whether the task is regression or classification

```python
print("Target (y) dtype:", y.dtype)
```

### Step 3. Identify the number of samples and columns of both the data matrix and the target

```python
print("X shape:", x.shape)   
print("y shape:", y.shape) 
```

### Step 4. Summarize the distribution of the data (min, max, median, mean, and standard deviation)

### Step 5. Plot the feature vs the target

```python
plt.scatter(x, y)
plt.xlabel('Experience Years')
plt.ylabel('Salary')
plt.title('Experience Years vs Salary')
plt.show()
```

### Step 7. Initialize a regression model

```python
model = LinearRegression()
model.fit(x,y)
```

### Step 8. Split the dataset into train and test sets

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
print("X_train shape:", X_train.shape)
print("y_train shape:", y_train.shape)
```

### Step 9.a Train the model on the training set

```python
model.fit(X_train, y_train)
```

### Step 9.b Identify the learned slope and the y-intercept of the linear regression model

```python
slope = model.coef_[0]
intercept = model.intercept_
print(f"Model parameters: Slope = {slope:.4f}, Intercept = {intercept:.4f}")
```

### Step 9.c how much salary does a person get with 2 years of experience?

```python
salary = model.predict([[2]])
print(f"Predicted salary for 2 years of experience: {salary[0]:.2f}")
```

### Step 9.d how much does an increase of 6 months of experience adds to a person's salary?

```python
salary_increase = model.coef_[0] * 0.5
print(f"Salary increase for 6 months of experience: {salary_increase:.2f}")
```

### Step 9.e if one has zero experience, then how much is the estimated salary?

```python
salary_zero_exp = model.intercept_
print(f"Estimated salary with zero experience: {salary_zero_exp:.2f}")
```

### Step 10. Evaluate the model on the test set

```python

y_pred = model.predict(X_test)

r2 = r2_score(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)

print("R2 Score:", r2)
print("MSE:", mse)
```

### Step 11. Take a sample of three rows from the test set and make a prediction on them and check whether they match the true label or not

```python
sample = X_test.sample(3, random_state=42)
sample_pred = model.predict(sample)
sample_true = y_test.loc[sample.index]

for i, (pred, true) in enumerate(zip(sample_pred, sample_true)):
    print(f"Sample {i+1}: Predicted={pred:.2f}, Actual={true:.2f}")
```

### Step 12. Plot the data and the regression line

```python
plt.scatter(X_test, y_test, label="Actual")
plt.plot(X_test, y_pred, color="red", label="Regression Line")
plt.xlabel("Experience Years")
plt.ylabel("Salary")
plt.title("Data vs Regression Line")
plt.legend()
plt.show()
```

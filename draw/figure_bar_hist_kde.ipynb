# Draw Figures

```python
import seaborn as sns

penguins = sns.load_dataset('penguins')
penguins.head()
```

```python
penguins[['body_mass_g']].describe()
```

```python
penguins[['sex']].value_counts()
```

```python
penguins.sample(3)
```

```python
sns.displot(
    data=penguins,
    kind='hist', # Histogram
    discrete=True, shrink=0.8, # Categorical
    x='species', # Variable to plot on the x-axis
    color='white',
    height=3, aspect=1 # Figure size
)
```

```python
sns.displot(
    data=penguins,
    kind='hist', # Histogram
    # discrete=True, shrink=0.8, # Categorical
    x='body_mass_g', # Variable to plot on the x-axis
    color='white',
    height=3, aspect=1 # Figure size
)
```

```python
sns.displot(
    data=penguins,
    kind='kde', # Histogram
    # discrete=True, shrink=0.8, # Categorical
    x='body_mass_g', # Variable to plot on the x-axis
    color='black',
    height=3, aspect=1, # Figure size
)
```

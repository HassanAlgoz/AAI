```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

# Generate x and y
mean_iq = 100
std = 13.2
x_min = 60
x_max = 140

x = np.linspace(x_min, x_max, 1000)
y = norm.pdf(x, loc=mean_iq, scale=std)

q1 = np.percentile(x, 25)
q3 = np.percentile(x, 75)

# Plot the curves matching the visual colors
fig, ax = plt.subplots(figsize=(6, 4))
ax.plot(x, y, color='black', linewidth=2)
plt.ylim(bottom=0)

ax.axvline(q1, color='firebrick', linestyle='--', linewidth=1)
ax.axvline(q3, color='firebrick', linestyle='--', linewidth=1)

ax.text(q1, ax.get_ylim()[1]*0.8, 'Q1', color='firebrick', ha='right', fontsize=12)
ax.text(q3, ax.get_ylim()[1]*0.8, 'Q3', color='firebrick', ha='left', fontsize=12)

# shade the area between q1 and q3
ax.fill_between(x, y, where=(x >= q1) & (x <= q3), color='firebrick', alpha=0.3)

# 5. Formatting to match the original image's clean style
ax.set_title('IQ Score Frequency Distribution', fontsize=14, fontweight='bold', pad=15)
ax.set_xlabel('IQ Score', fontsize=12, fontweight='bold', labelpad=10)

# Set the specific x-axis tick marks
ax.set_xticks(range(x_min, x_max+1, 10))

fig.tight_layout()
```

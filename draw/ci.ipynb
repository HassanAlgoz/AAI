```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Setup the x-axis (representing percentages, centered at 50)
x = np.linspace(0, 60, 1000)
mean = 20

# Define the two distributions (smaller std dev = taller/narrower peak)
y_tall = stats.norm.pdf(x, mean, 2)  # High sample size, low variance
y_short = stats.norm.pdf(x, mean, 8) # Low sample size, high variance

# Create the plot
fig, ax = plt.figure(figsize=(10, 6)), plt.gca()

# Plot the curves
ax.plot(x, y_tall, color='teal', linewidth=2.5)
ax.plot(x, y_short, color='coral', linewidth=2.5)

# Pin the annotations directly to the peaks
# xytext=(15, 0) forces the text to sit exactly 15 points to the right of the peak
ax.annotate(
    "Sample size = 3000\nMargin of error: 2%", 
    xy=(mean, y_tall.max()), xycoords='data',
    xytext=(15, 0), textcoords='offset points', 
    ha='left', va='center', fontsize=12, color='teal', fontweight='bold',
    arrowprops=dict(arrowstyle="-", color='teal')
)

ax.annotate(
    "Sample size = 30\nMargin of error: 10%", 
    xy=(mean, y_short.max()), xycoords='data',
    xytext=(15, 10), textcoords='offset points', 
    ha='left', va='center', fontsize=12, color='coral', fontweight='bold',
    arrowprops=dict(arrowstyle="-", color='coral')
)

# Clean up visual clutter
ax.set_yticks([]) 
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)

plt.tight_layout()
plt.show()
```

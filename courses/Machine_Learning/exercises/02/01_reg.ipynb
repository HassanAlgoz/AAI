## What is Linear Regression?

[`LinearRegression`](https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LinearRegression.html) fits the **straight line** that best matches your data:

$$y = m \cdot x + b$$

- **slope** $m$ (`model.coef_`): how much the target changes for **each one-unit** increase in the feature.
- **intercept** $b$ (`model.intercept_`): the predicted target when the feature is **0**.
- After fitting, `model.score(X, y)` reports $R^2$ — the fraction of variance explained (1.0 is perfect).

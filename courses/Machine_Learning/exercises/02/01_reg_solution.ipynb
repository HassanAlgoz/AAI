## What is Linear Regression?

[`LinearRegression`](https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LinearRegression.html) fits the **straight line** that best matches your data:

$$y = m \cdot x + b$$

- **slope** $m$ (`model.coef_`): how much the target changes for **each one-unit** increase in the feature.
- **intercept** $b$ (`model.intercept_`): the predicted target when the feature is **0**.
- After fitting, `model.score(X, y)` reports $R^2$ — the fraction of variance explained (1.0 is perfect).


## Setup

```python
# https://pandas.pydata.org/docs/
import pandas as pd

# https://numpy.org/doc/
import numpy as np

# https://seaborn.pydata.org/
import seaborn as sns

# https://matplotlib.org/
import matplotlib.pyplot as plt

# https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LinearRegression.html
from sklearn.linear_model import LinearRegression

# https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.train_test_split.html
from sklearn.model_selection import train_test_split

DATA_URL = "https://raw.githubusercontent.com/HassanAlgoz/AAI/main/courses/Machine_Learning/datasets/Rounded_Student_Hours_Studied_vs_Marks_Dataset.csv"
```

## Step 1 — Load the data

Read the CSV at `DATA_URL` into a DataFrame and take a first look.


### Step 1. Load the data

```python
df = pd.read_csv(DATA_URL)
df.head()
```

## Step 2 — Features and target

Assign **`X`** to the feature(s) and **`y`** to the target, then inspect their types.

**Predict:** will the target be whole categories or continuous numbers?


### Step 2.a Assign variables `X` to the features and `y` to the target

```python
X = df[["Hours_Studied"]]
y = df["Marks"]
```

### Step 2.b Print the type of each

```python
display(X.dtypes)
display(y.dtype)
```

### Step 2.c Is the target categorical or numerical? Is this regression or classification?

- Target is **numerical** (marks are continuous numbers).
- Task is **regression** because the target is numerical.


## Step 3 — Dataset shape

How many samples and columns are in `X` and `y`?


### Step 3. Identify the number of samples and columns of the data matrix and the target

```python
print("X shape:", X.shape)
print("y shape:", y.shape)
```

## Step 4 — Summary statistics

Summarize the distribution (min, max, median, mean, standard deviation).


### Step 4. Summarize the distribution of the data

```python
df.describe()
```

## Step 5 — Plot the feature vs the target

Draw a scatter plot of the feature against the target.

**Predict:** do you expect the points to trend up, down, or show no pattern?


### Step 5. Plot the feature vs the target

```python
fig, ax = plt.subplots(figsize=(6, 4))

sns.scatterplot(
    data=df,
    x="Hours_Studied",
    y="Marks",
    ax=ax,
)
ax.set_title("Hours studied vs marks")
fig.tight_layout()
plt.show()
```

### Step 6. What is the relationship between the feature and the target? (increasing / decreasing / none)

**Increasing.** As hours studied go up, marks tend to go up too.


## Step 7 — Initialize the model

Create a `LinearRegression` instance.


### Step 7. Initialize a `LinearRegression` model

```python
model = LinearRegression()
```

## Step 8 — Train/test split

Hold out 20% of the data for testing. Use `random_state=10` for a reproducible split.


### Step 8. Split the dataset into train and test sets

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=10,  # reproducible split
)
```

## Step 9 — Train and interpret

### Step 9.a Train the model on the training set

```python
model.fit(X_train, y_train)
```

### Step 9.b Identify the learned slope and the y-intercept of the model

```python
slope = model.coef_[0]
intercept = model.intercept_

print(f"slope (marks per hour): {slope:.2f}")
print(f"intercept (marks at 0 hours): {intercept:.2f}")
```

### Step 9.c How many marks do we gain per 1 hour of study?

**Predict:** roughly how many marks per hour, based on the scatter plot?


```python
print(f"Each extra hour of study adds about {slope:.2f} marks.")
```

### Step 9.d How much does a 10-minute increase of study time add to our marks?

```python
ten_minutes = 10 / 60  # express 10 minutes in hours
added = ten_minutes * slope
print(f"10 extra minutes of study adds about {added:.2f} marks.")
```

### Step 9.e If one doesn't study at all, how many marks do they get?

```python
print(f"With no study at all, the model predicts about {intercept:.2f} marks.")
```

## Step 10 — Evaluate on the test set

Report the model's $R^2$ score on the held-out test set.

**Predict:** will the score be close to 1.0, or noticeably lower?


### Step 10. Evaluate the model on the test set

```python
r2 = model.score(X_test, y_test)
print(f"R^2 on the test set: {r2:.3f}")
```

## Step 11 — Spot-check predictions

Take three rows from the test set, predict their marks, and compare against the true marks.


### Step 11. Predict on three test rows and compare to the true labels

```python
sample_features = X_test.head(3)
sample_truth = y_test.head(3)

pd.DataFrame(
    {
        "Hours_Studied": sample_features["Hours_Studied"].to_numpy(),
        "predicted_marks": model.predict(sample_features).round(2),
        "actual_marks": sample_truth.to_numpy(),
    }
)
```

## Step 12 — Plot the regression line

Overlay the learned line on the scatter plot.

**Predict:** will the line be straight? Why or why not?


### Step 12. Plot the data and the regression line

```python
x_grid = np.linspace(
    X["Hours_Studied"].min(),
    X["Hours_Studied"].max(),
    num=100,
)
line_df = pd.DataFrame({"Hours_Studied": x_grid})
line_df["Marks"] = model.predict(line_df)

fig, ax = plt.subplots(figsize=(6, 4))

sns.scatterplot(
    data=df,
    x="Hours_Studied",
    y="Marks",
    color="black",
    alpha=0.5,
    ax=ax,
)
sns.lineplot(
    data=line_df,
    x="Hours_Studied",
    y="Marks",
    color="red",
    ax=ax,
)
ax.set_title("Linear regression fit")
fig.tight_layout()
plt.show()
```

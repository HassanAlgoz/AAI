## Setup

```python
# https://pandas.pydata.org/docs/
import pandas as pd

# https://numpy.org/doc/
import numpy as np

# https://seaborn.pydata.org/
import seaborn as sns

# https://matplotlib.org/
import matplotlib.pyplot as plt

# https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LinearRegression.html
from sklearn.linear_model import LinearRegression

# https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.train_test_split.html
from sklearn.model_selection import train_test_split

DATA_URL = "https://raw.githubusercontent.com/HassanAlgoz/AAI/main/courses/Machine_Learning/datasets/Salary%20Data.csv"
```

## Challenge 1 — Load and explore

1. Read the CSV at `DATA_URL`.
2. Assign `X` to `Experience Years` and `y` to `Salary`.
3. Check the dtypes, the shape, and `describe()`.
4. Scatter-plot experience against salary.

**Predict:** which way will the points trend, and is this regression or classification?


### Challenge 1.a Load and assign `X` / `y`

```python
df = pd.read_csv(DATA_URL)

X = df[["Experience Years"]]
y = df["Salary"]

df.head()
```

### Challenge 1.b Explore shape, summary stats, and a scatter plot

```python
print("X shape:", X.shape)
print("y shape:", y.shape)
display(df.describe())

fig, ax = plt.subplots(figsize=(6, 4))
sns.scatterplot(
    data=df,
    x="Experience Years",
    y="Salary",
    ax=ax,
)
ax.set_title("Experience vs salary")
fig.tight_layout()
plt.show()
```

## Challenge 2 — Fit a linear model

Initialize `LinearRegression`, split into train/test (20% test, `random_state=10`), fit, then read the **slope** and **intercept**.


### Challenge 2. Initialize, split, fit, and read the coefficients

```python
model = LinearRegression()

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=10,  # reproducible split
)
model.fit(X_train, y_train)

slope = model.coef_[0]
intercept = model.intercept_

print(f"slope (salary per year): {slope:,.2f}")
print(f"intercept (salary at 0 years): {intercept:,.2f}")
```

## Challenge 3 — Interpret the coefficients

Use the slope and intercept to answer:

1. What salary does the model predict for **2 years** of experience?
2. How much does an extra **6 months** of experience add?
3. What salary does it predict at **zero** experience?


### Challenge 3. Turn the coefficients into plain-English answers

```python
salary_2_years = model.predict(pd.DataFrame({"Experience Years": [2]}))[0]
print(f"Salary at 2 years of experience: {salary_2_years:,.2f}")

six_months = 0.5  # express 6 months in years
added = six_months * slope
print(f"An extra 6 months of experience adds about {added:,.2f}")

print(f"Salary at zero experience: {intercept:,.2f}")
```

## Challenge 4 — Evaluate and visualize

1. Report the $R^2$ score on the test set.
2. Spot-check predictions on three test rows.
3. Plot the regression line over the scatter plot.

**Predict:** will the fit look tighter or looser than the Hours vs Marks data?


### Challenge 4.a Score and spot-check

```python
r2 = model.score(X_test, y_test)
print(f"R^2 on the test set: {r2:.3f}")

sample_features = X_test.head(3)
sample_truth = y_test.head(3)

pd.DataFrame(
    {
        "Experience Years": sample_features["Experience Years"].to_numpy(),
        "predicted_salary": model.predict(sample_features).round(2),
        "actual_salary": sample_truth.to_numpy(),
    }
)
```

### Challenge 4.b Plot the regression line

```python
x_grid = np.linspace(
    X["Experience Years"].min(),
    X["Experience Years"].max(),
    num=100,
)
line_df = pd.DataFrame({"Experience Years": x_grid})
line_df["Salary"] = model.predict(line_df)

fig, ax = plt.subplots(figsize=(6, 4))

sns.scatterplot(
    data=df,
    x="Experience Years",
    y="Salary",
    color="black",
    alpha=0.5,
    ax=ax,
)
sns.lineplot(
    data=line_df,
    x="Experience Years",
    y="Salary",
    color="red",
    ax=ax,
)
ax.set_title("Linear regression fit")
fig.tight_layout()
plt.show()
```

# Pandas: DataFrame and Series

**Pandas** provides fast, flexible, and expressive data structures designed to make working with “relational” or “labeled” data both **easy and intuitive**. Pandas is also **very fast** since it builds on top of **NumPy**, which builds on top of **C**.

The community agreed alias for `pandas` is `pd`, so it has been standard practice:

````python
import pandas as pd
````

## Read data in

Data can be loaded into a Pandas DataFrame from various sources and different formats:

1. **Databases**: relational (SQL) and unstructured (NoSQL)
2. **Surveys**: Google Forms, SurveyMonkey
3. **APIs**: Social Media Platforms, Open Street Maps, Flights, ..etc.
4. **Web Scraping**: BeautifulSoup, Scrapy, Selenium
5. **Raw Files**: csv, excel, txt, json, xml, pdf, ..etc.

Pandas provides the [`read_csv()`](https://pandas.pydata.org/docs/reference/api/pandas.read_csv.html#pandas.read_csv "pandas.read_csv") function to read data stored as a csv file into a pandas `DataFrame`. pandas supports many different file formats or data sources out of the box (csv, excel, sql, json, parquet, …), each of them with the prefix `read_*`.

![](../assets/pandas/pandas_io.png)

See: [IO tools (text, CSV, HDF5, …)](https://pandas.pydata.org/docs/user_guide/io.html) for a table containing available readers and writers.

````python
help(pd.read_csv)
````

The result of a `read_*` operation is a `DataFrame` (usually abbreviated as `df`), which is essentially a **table of rows and columns**.

![Figure: Pandas' DataFrame](../assets/pandas/dataframe.png)

Let's demonstrate that.

### read from file on disk into a DataFrame

````python
# Best practice: use pathlib to handle file paths
from pathlib import Path

# Set the path to the directory containing the datasets
data_dir = Path('../data/')

# Debugging
print(data_dir.exists())
````

````python
print(data_dir.resolve())
````

````python
file_path = data_dir / 'abc.csv'

# Debugging
print(file_path.exists())
````

Once we have specified the path, we load the data into a Python object; i.e., a `DataFrame`:

````python
# Notice we use `df` to denote a DataFrame
df_1 = pd.read_csv(file_path)

# Debugging
df_1.head() # show first 5 rows
````

### Modify

````python
df_1 = df_1 * 10
df_1
````

### Write the DataFrame to csv file on disk

````python
df_1.to_csv(data_dir / 'abc_2.csv', index=False)
````

### read from url

Alternatively, one could load the data from a url directly:

````python
# GitHub URL (replace "blob" with "raw" to get the raw content)
data_url_wpm = "https://github.com/HassanAlgoz/B5/blob/next/courses/Data_Wrangling/data/wpm.csv"
data_url_wpm = data_url_wpm.replace("blob", "raw")
df_2 = pd.read_csv(data_url_wpm)

# Debugging
df_2.tail() # show last 5 rows
````

## DataFrame Metadata

The `head`/`tail`/`info` methods and the `dtypes` attribute are convenient for a first check:

````python
# show only data types
df_2.dtypes
````

````python
# show full technical details of the DataFrame
df_2.info()
````

## Creating a DataFrame

Sometimes, like in demonstration, you need to create a `DataFrame` object, not loaded from outside.

DataFrames can be constructed from Python dictionaries, like so:

````python
df = pd.DataFrame(
    {
        'A': [10, 20, 30],
        'B': [40, 50, 60]
    }
)

df
````

- Keys `'A'` and `'B'` become column labels
- Values become cell values

## Key Takeaways

- **DataFrames** are core to pandas: they provide labeled, 2-dimensional data structures, allowing flexible and powerful manipulation and analysis of tabular data.

- **Reading Data:** Easily load data from various sources (CSV, Excel, SQL, web, etc.) with functions like `pd.read_csv()`.

- **Quick Inspection:** Use methods such as `.head()`, `.tail()`, `.info()`, and `.dtypes` to quickly get an overview of your data.
